```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

X = np.linspace(0.1,10,100)
y = X**(2) - np.log(X)*3 + X*4 - np.cos(2*np.pi*X/8.8)*100 + np.random.rand(100)*np.sqrt(X)*52
plt.scatter(X, y)
plt.show()
```

```python
plt.figure(figsize=(15,20))
lrT = LinearRegression()
lrT.fit(X.reshape((-1,1)), y)

X_p = np.linspace(0,10,1000)

plt.subplot('421')
plt.scatter(X,y)
plt.plot(X_p, lrT.predict(X_p.reshape((-1,1))).flatten(), c='r', label='1st degree')
plt.title('MSE is %.5f' % mean_squared_error(y, lrT.predict(X.reshape((-1,1)))))
plt.legend()

pf = PolynomialFeatures(degree=2)
lrT.fit(pf.fit_transform(X.reshape(-1,1)), y)

plt.subplot('422')
plt.scatter(X,y)
plt.plot(X_p, lrT.predict(pf.transform(X_p.reshape((-1,1)))), c='b', label='2nd degree')
plt.title('MSE is %.5f' % mean_squared_error(y, lrT.predict(pf.transform(X.reshape((-1,1))))))
plt.legend()

pf = PolynomialFeatures(degree=3)
lrT.fit(pf.fit_transform(X.reshape(-1,1)), y)

plt.subplot('423')
plt.scatter(X,y)
plt.plot(X_p, lrT.predict(pf.transform(X_p.reshape((-1,1)))), c='g', label='3rd degree')
plt.title('MSE is %.5f' % mean_squared_error(y, lrT.predict(pf.transform(X.reshape((-1,1))))))
plt.legend()

pf = PolynomialFeatures(degree=4)
lrT.fit(pf.fit_transform(X.reshape(-1,1)), y)

plt.subplot('424')
plt.scatter(X,y)
plt.plot(X_p, lrT.predict(pf.transform(X_p.reshape((-1,1)))), c='#000333', label='4th degree')
plt.title('MSE is %.5f' % mean_squared_error(y, lrT.predict(pf.transform(X.reshape((-1,1))))))
plt.legend()

pf = PolynomialFeatures(degree=5)
lrT.fit(pf.fit_transform(X.reshape(-1,1)), y)

plt.subplot('425')
plt.scatter(X,y)
plt.plot(X_p, lrT.predict(pf.transform(X_p.reshape((-1,1)))), c='#00eeee', label='5th degree')
plt.title('MSE is %.5f' % mean_squared_error(y, lrT.predict(pf.transform(X.reshape((-1,1))))))
plt.legend()

pf = PolynomialFeatures(degree=6)
lrT.fit(pf.fit_transform(X.reshape(-1,1)), y)

plt.subplot('426')
plt.scatter(X,y)
plt.plot(X_p, lrT.predict(pf.transform(X_p.reshape((-1,1)))), c='#000aa3', label='6th degree')
plt.title('MSE is %.5f' % mean_squared_error(y, lrT.predict(pf.transform(X.reshape((-1,1))))))
plt.legend()

pf = PolynomialFeatures(degree=7)
lrT.fit(pf.fit_transform(X.reshape(-1,1)), y)

plt.subplot('427')
plt.scatter(X,y)
plt.plot(X_p, lrT.predict(pf.transform(X_p.reshape((-1,1)))), c='#0dd333', label='7th degree')
plt.title('MSE is %.5f' % mean_squared_error(y, lrT.predict(pf.transform(X.reshape((-1,1))))))
plt.legend()

pf = PolynomialFeatures(degree=15)
lrT.fit(pf.fit_transform(X.reshape(-1,1)), y)

plt.subplot('428')
plt.scatter(X,y)
plt.plot(X_p, lrT.predict(pf.transform(X_p.reshape((-1,1)))), c='#bb0333', label='15th degree')
plt.title('MSE is %.5f' % mean_squared_error(y, lrT.predict(pf.transform(X.reshape((-1,1))))))
plt.legend()
plt.suptitle('Affect of polynomial features on MSE')
plt.tight_layout(pad=6)
plt.show()
```
